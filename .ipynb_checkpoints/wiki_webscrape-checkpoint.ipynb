```python
from bs4 import BeautifulSoup
import lxml
import requests
import pandas as pd
```

```python
html = "https://en.wikipedia.org/wiki/Night"
```

```python
r = requests.get(html)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
soup
```

```python
t1 = soup.body
t1
```

```python
t1.attrs
```

```python
o = soup.find("div", class_ = "mw-content-container")
o
```

```python
o.header.h1.text
```

```python
t = o.find("div", class_ = "thumb tright")
t
```

```python
tt = t.find_all("a")
tt
```

```python
lnk = []
for i in tt:
    fllink = i.get("href")
    lnk.append(fllink)
```

```python
lnkk = lnk[0]
print(type(lnkk))
```

```python
lnkk
```

```python
lnkkhttp = "https://en.wikipedia.org"
```

```python
full_link = lnkkhttp + lnkk
full_link
```
