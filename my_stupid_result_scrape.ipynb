```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
```

```python
data = requests.get("http://app1.helwan.edu.eg/Science/HasasnUpMlist.asp?z_dep=%3D&z_st_name=LIKE&z_st_settingno=%3D&x_st_settingno=&x_st_name=&z_gro=%3D&x_gro=%C7%E1%CB%C7%E1%CB%C9&x_dep=%C7%E1%D1%ED%C7%D6%ED%C7%CA+%E6+%C7%E1%DD%ED%D2%ED%C7%C1+%E6+%C7%E1%DD%E1%DF&z_sec=LIKE&x_sec=%C7%E1%C7%CD%D5%C7%C1+%E6%DA%E1%E6%E3+%C7%E1%CD%C7%D3%C8&psearch=&Submit=++++%C8%CD%CB++++").text
```

```python
soup = BeautifulSoup(data, "lxml")
```

```python
soup.prettify()
```

```python
students_table = soup.find("table", id= "ewlistmain")
print(students_table)
```

```python
name = students_table.find(name= ["td", "st_name"])
# mark = students_table.find(name= ["td", "TNesba"]).text
sname = name.text
print(sname)
print()
# print(mark)
```

```python
names = []
for name in students_table.find_all(name= ["td", "st_name"]):
    names.append(name.text)
    print(names)
```

```python
names_df = pd.DataFrame(names, columns=["Name"])
names_df
```
